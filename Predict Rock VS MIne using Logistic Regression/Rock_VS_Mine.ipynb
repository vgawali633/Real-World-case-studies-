```python
#import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv("/content/sonar data.csv",header=None)
# here we took header=None because there is no any header
```

```python
"""split data in to X and y
   here we see that 61 columns out of which we took 1st 60 coumns in X and 
   last column in y as we want to predict the output...axis=1 means columns and axis =0 means row
   """

X=df.drop(columns=60,axis=1)
y=df[60]
```

```python
#train test split 

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
#Bulid the model and find the accuracy
# for fiting the data into model we always use X_train and y_train

lr=LogisticRegression()
lr.fit(X_train,y_train) 
```

```python
#predict the model
#for predict the model we use X_test only
pred=lr.predict(X_test)
pred
```

```python
#find Accuracy of the model
acc=lr.score(X_test,y_test)
print("The Model Accuracy is",acc) #so your model gives 77 % accuracy
```
